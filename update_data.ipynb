```python
from kaggle.api.kaggle_api_extended import KaggleApi
from pathlib import Path
import zipfile
import os
def download_latest():
    api = KaggleApi()
    api.authenticate()
    script_dir = Path(os.getcwd())
    for fname in ["Games.csv", "PlayerStatistics.csv"]:
        api.dataset_download_file("eoinamoore/historical-nba-data-and-player-box-scores",fname,path=str(script_dir),force=True,quiet=False,)
        zip_path = script_dir / f"{fname}.zip"
        if zip_path.exists():
            with zipfile.ZipFile(zip_path, "r") as z:
                z.extractall(script_dir)
            zip_path.unlink()
    games_file = script_dir/"Games.csv"
    new_name = script_dir/"Games(1).csv"
    if games_file.exists():
        if new_name.exists():
            new_name.unlink()
        games_file.rename(new_name)
download_latest()
```
